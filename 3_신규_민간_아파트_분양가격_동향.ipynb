##2021데이터 전처리

###결측치

```python
#각 특성별 NaN의 개수
df_2021.isnull().sum()
df_2021.isna().sum() #두 개 같은거
```

```python
#NaN은 float이기에 위의 다른 숫자도 float로 처리됨
df_2021["분양가격"] = df_2021["분양가격"].dropna()
df_2021
#분양가격에 dropna를 걸어주면 전체 DataFrame에서 제거되지 않음
```

```python
#이렇게 데이터프레임에 걸어줘야 NaN을 제거 할 수 있음
df_2021=df_2021.dropna()#axis=1을 넣으면 NaN이 있는 열을 지워버림 조심!
df_2021.info()
```

###데이터 변환

```python
#pd.to_numeric(df_2021["분양가격(㎡)"])#밑에 함수를 그냥 실행하면 error가 나옴." "이 포함되어 있기 때문
#df_2021["분양가격(㎡)"].astype(int)
#그렇기에 위의 함수로 numeric으로 바꾸어 줘야함. 하지만 위의 함수도 error나옴
df_2021["분양가격"]=pd.to_numeric(df_2021["분양가격(㎡)"],errors='coerce')#errors=ignore면 그냥 무시 coerce면 NaN으로 반환
```

```python
#2013데이터는 평당 가격이기에 위의 2021의 데이터의 단위를 변경해야함(그래서 float형식으로 변경)
df_2021["평당분양가격"]=df_2021["분양가격"]*3.3
df_2021
```

```python
df_2021["전용면적"]=df_2021["규모구분"].str.replace("전용면적","")
df_2021["전용면적"]=df_2021["전용면적"].str.replace("초과","~")
df_2021["전용면적"]=df_2021["전용면적"].str.replace("이하","")
df_2021["전용면적"]=df_2021["전용면적"].str.replace(" ","").str.strip()
df_2021["전용면적"]=df_2021["전용면적"].str.replace("모든면적","전체")
df_2021
```

```python
df_2021.info()
```

```python
df_2021=df_2021.drop(["규모구분","분양가격(㎡)"],axis=1)
```

##2013데이터 전처리

```python
df_2013.head()
#지역명,연도,월,평당분양가격이 필요
```

```python
df_2013_melt=pd.melt(df_2013,id_vars="지역",var_name="기간",value_name="평당분양가격")
df_2013_melt.columns=["지역명","기간",'평당분양가격']
df_2013_melt
```

```python
def get_year(date):
  year = date.split("년")[0]
  year = int(year)
  return year

print(get_year("2021년4월"))
```

```python
def get_month(date):
  month = date.split("년")[1].replace("월","")
  month = int(month)
  return month

print(get_month("2021년4월"))
```

```python
df_2013_melt["연도"]=df_2013_melt["기간"].apply(get_year)
df_2013_melt.head()
```

```python
df_2013_melt["월"]=df_2013_melt["기간"].apply(get_month)
df_2013_melt.head()
```

```python
#df_2013_melt=df_2013_melt.drop(["기간"],axis=1)
df_2013_melt.head()
```
